```python
srcs=[]
targets=[]

with open("data/squad-src-train.txt","r")as f:
    for line in f:
        srcs.append(line.strip())
with open("data/squad-tgt-train.txt","r")as f:
    for line in f:
        targets.append(line.strip())
```

```python
for t in srcs[0:10]:
    print(t)
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

counter=0
for s,t in zip(srcs,targets):
    check=False
    for w in t.split():
        if s.find(w)!=-1 and w not in stop_words:
            check=True
            break
    if check==False:
        print(s)
        print(t)
        print()
        counter+=1
```
